```python
import matplotlib.pyplot as plt
import numpy as np

while (True):
    print("please input with format : a,b,c,d and all need to be a number")
    inputStr = input()
    strlist = inputStr.split(',')
    if (strlist.__len__() != 4):
        continue
    try:
        a = int(strlist[0])
        b = int(strlist[1])
        c = int(strlist[2])
        d = int(strlist[3])
    except:
        print("please check the format and enter again.")
        continue
    else:
        break
print("please input with format : ax,ay,bx,by,cx,cy... and all need to be a number")
fivePtXYarray = []
while (True):
    fivePtStr = input()
    fivePtStrlist = fivePtStr.split(',')
    if (fivePtStrlist.__len__() % 2 != 0):
        print("Error! please check the format")
        continue
    try:
        for i in range(0, fivePtStrlist.__len__()):
            fivePtXYarray.append(int(fivePtStrlist[i]))
    except:
        print("Error! please check the format")
        fivePtXYarray.clear()
        continue
    else:
        break


X = np.linspace(-25, 25, 100)

Y = a * (X ** 3) + b * (X ** 2) + c * (X ** 1) + d
plt.plot(X, Y, color="purple", linewidth=1.0, linestyle="-",
         label="%dx^3+%dx^2+%dx+%d" % (a, b, c, d))



redFlag = False
greenFlag = False
blueFlag = False
for i in range(0, int(fivePtStrlist.__len__() / 2.0)):
    pt_x = fivePtXYarray[2 * i]
    pt_y = fivePtXYarray[2 * i + 1]
    eq_y = a * (pt_x ** 3) + b * (pt_x ** 2) + c * (pt_x ** 1) + d

    if pt_y < eq_y:
        if (blueFlag):
            plt.plot(pt_x, pt_y, color="blue", marker="o")
        else:
            plt.plot(pt_x, pt_y, color="blue", marker="o", label="lower")
            blueFlag = True
    elif pt_y > eq_y:
        if (redFlag):
            plt.plot(pt_x, pt_y, color="red", marker="o")
        else:
            plt.plot(pt_x, pt_y, color="red", marker="o", label="higher")
            redFlag = True
    else:
        if (greenFlag):
            plt.plot(pt_x, pt_y, color="green", marker="o")
        else:
            plt.plot(pt_x, pt_y, color="green", marker="o", label="equal")
            greenFlag = True

plt.legend(bbox_to_anchor=(1, 1), loc=1, borderaxespad=0.)
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.show()
print("this picture corresponds to Q3 and Q4.")


now_x1 = 0
now_x2 = 10
gamma = 0.01
for i in range (0,1000):
    now_m1 = a * 3*(now_x1 ** 2) + b * 2*now_x1 + c
    now_m2 = a * 3*(now_x2 ** 2) + b * 2*now_x2 + c
    if(0 <= now_x1 - gamma*now_m1 <= 10):
        now_x1 -= (gamma*now_m1)
    if(0 <= now_x2 - gamma*now_m2 <= 10):
        now_x2 -= (gamma*now_m2)

now_y1 = a * (now_x1 ** 3) + b * (now_x1 ** 2) + c * (now_x1 ** 1) + d
now_y2 = a * (now_x2 ** 3) + b * (now_x2 ** 2) + c * (now_x2 ** 1) + d
if(now_y1 < now_y1):
    print("min of this function between X=0 and X=10 is ",now_y1)
else:
    print("min of this function between X=0 and X=10 is ",now_y2)

now_x1 = 0
now_x2 = 10
for i in range (0,1000):
    now_m1 = a * 3*(now_x1 ** 2) + b * 2*now_x1 + c
    now_m2 = a * 3*(now_x2 ** 2) + b * 2*now_x2 + c
    if(0 <= now_x1 + gamma*now_m1 <= 10):
        now_x1 += (gamma*now_m1)
    if(0 <= now_x2 + gamma*now_m2 <= 10):
        now_x2 += (gamma*now_m2)

now_y1 = a * (now_x1 ** 3) + b * (now_x1 ** 2) + c * (now_x1 ** 1) + d
now_y2 = a * (now_x2 ** 3) + b * (now_x2 ** 2) + c * (now_x2 ** 1) + d
if(now_y1 > now_y1):
    print("max of this function between X=0 and X=10 is ",now_y1)
else:
    print("max of this function between X=0 and X=10 is ",now_y2)

print("Above result corresponds to Q5.")

```
